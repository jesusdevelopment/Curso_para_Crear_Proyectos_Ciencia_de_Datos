## Importación de Bibliotecas

```python
import pandas as pd
import os
```

## Carga de Láminas del archivo Excel

```python
df_contratistas= pd.read_excel(io='../data/12_DICIEMBRE_2020.xlsx', sheet_name='CONTRATISTAS Y FDO FED', skiprows=5 )
df_contratistas['SHEET'] = 'CONTRATISTAS Y FDO FED'
df_contratistas.head()
```

```python
df_gastos_varios= pd.read_excel(io='../data/12_DICIEMBRE_2020.xlsx', sheet_name='GASTOS VARIOS', skiprows=5 )
df_gastos_varios['SHEET'] = 'GASTOS VARIOS'
df_gastos_varios.head()
```

```python
df_servicios= pd.read_excel(io='../data/12_DICIEMBRE_2020.xlsx', sheet_name='SERV PPROF', skiprows=5 )
df_servicios['SHEET'] = 'SERV PPROF'
df_servicios.head()
```

```python
df_comunicacions= pd.read_excel(io='../data/12_DICIEMBRE_2020.xlsx', sheet_name='COMUNICACION', skiprows=5 )
df_comunicacions['SHEET'] = 'COMUNICACION'
df_comunicacions.head()
```

```python
df_personales= pd.read_excel(io='../data/12_DICIEMBRE_2020.xlsx', sheet_name='SERV. PERS.', skiprows=5 )
df_personales['SHEET'] = 'SERV. PERS.'
df_personales.head()
```

```python
df_compras_dic= pd.concat([df_contratistas, df_gastos_varios, df_servicios, df_comunicacions, df_personales], axis=0, ignore_index=True)
df_compras_dic.head()
```

## Eliminación de Filas con Valores Nulos

```python
df_compras_dic=df_compras_dic.dropna()
df_compras_dic.shape
```

```python
df_compras_dic.info()
```

## Corroboración de valores de columna 'FECHA'

```python
df_compras_dic['FECHA'].apply(lambda x: x.month).value_counts()
```

## Creación de archivo CSV limpio

```python
df_compras_dic.to_csv('../data/clean_compras.csv', index=None)
```
