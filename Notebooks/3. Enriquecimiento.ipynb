## Importación de Bibliotecas

```python
import pandas as pd
import numpy as np
```

## Carga de dataSet

```python
df_compras=pd.read_csv('../data/clean_compras.csv')
df_compras.head()   
```

```python
df_compras.info()
```

## Expansión de Dia, Mes y Fecha

```python
df_compras['FECHA']= pd.to_datetime(df_compras['FECHA'], format='%Y-%m-%d')
df_compras.info()
```

```python
df_compras['AÑO']= df_compras['FECHA'].dt.year
df_compras['MES']= df_compras['FECHA'].dt.month
df_compras['DIA']= df_compras['FECHA'].dt.day
df_compras['DIA_SEMANA']= df_compras['FECHA'].dt.dayofweek
df_compras.head()
```

## Nuevas Variables Secundarias

```python
num_ventas_beneficiario=df_compras['BENEFICIARIO'].value_counts()
num_ventas_beneficiario.head(10)
```

```python
num_ventas_beneficiario=df_compras['BENEFICIARIO'].value_counts().to_dict()
num_ventas_beneficiario
list(num_ventas_beneficiario.keys())[:10]
```

```python
df_compras['BENEFICIARIO_SALES'] = df_compras['BENEFICIARIO'].map(num_ventas_beneficiario)
df_compras.head()
```

```python
media_importes_tipos_egresos= df_compras.groupby('SHEET')['IMPORTE'].mean().sort_values(ascending=False).reset_index().iloc[0:10]
media_importes_tipos_egresos.head(10)
```
